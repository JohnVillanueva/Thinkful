```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exponentialvar = np.random.exponential(1, 100)
plt.hist(exponentialvar, bins = 20, color = "c")
plt.axvline(exponentialvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponentialvar.mean() - exponentialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponentialvar.mean() + exponentialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
print("Mean: ",exponentialvar.mean())
print("Std Dev: ",exponentialvar.std())
```

The mean for a exponential distribution with scale 1 is ~1 while the standard deviation is also ~1. The mean is a useful figure because while the range of numbers from the graph shows because 0 and presumably positive infinite, there's clearly a steep drop off of numbers, and that the percentage of data points past 7 is negligible. The mean demonstrates that he concentration of numbers exists around 1. The standard deviation is not very useful because this distribution is far from normal. It is not symetric, so one standard deviation below the mean captures far more data points than one standard deviation above the mean. Like wise, past 2 deviations below the mean, there in no more data collect where as the graph shows that there is data to be captured even 6 stadard deviations above the mean.

```python
laplacevar = np.random.laplace(0, 1, 1000)
plt.hist(laplacevar, bins = 20, color = "c")
plt.axvline(laplacevar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(laplacevar.mean() - laplacevar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(laplacevar.mean() + laplacevar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
print("Mean: ",laplacevar.mean())
print("Std Dev: ",laplacevar.std())
```

The laplace distribution is essentially an exponential distribution but in both the positive and negative direction so that it creates a symetric distribution. Because it's symetric and the data tapers off in both the positive and negative directions, the data is considered normal. The both the mean and standard deviation have useful meanings in the context of this distribution.

In general, the standard deviation statistic is useful primarily when the data it describes is symetrically distributed and if the data tapers off in both directions. Most of the distributions don't have symetric distributions of data, while distribution liket the uniform distribution is is symetrica but does not taper off.

```python
norm1 = np.random.normal(5,0.5,1000)
norm2 = np.random.normal(10,1,1000)
norm3 = norm1 + norm2
plt.hist(norm3, bins = 20, color = 'c')
plt.axvline(norm3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(norm3.mean() - norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean() + norm3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
print("Mean: ",norm3.mean())
print("Std Dev: ",norm3.std())
print("Mode: ",np.median(norm3))
```
