```python
#Total Output per Year by Country
CountryOutput = faodf.groupby('Area').sum().drop(['Area Code', 'Item Code', 'Element Code', 'latitude', 'longitude'], axis=1)
CountryOutput.head()
```

```python
#Total Out Per Year by Food Industry
IndustryOutput = faodf.groupby('Item').sum().drop(['Area Code', 'Item Code', 'Element Code', 'latitude', 'longitude'], axis=1)
IndustryOutput.head()
```

```python
#What are the largest item industries by mass in 2013?
IndustryOutput['Y2013'].sort_values(ascending=False).head()
```

```python
#What are the smallest item industries by mass in 2013?
IndustryOutput['Y2013'].sort_values(ascending=False).tail()
```

```python
#Histogram of Item Industry Output Mass in 2013
HistItemOutput2013 = IndustryOutput['Y2013'].replace(0,np.nan).dropna()
plt.hist(HistItemOutput2013)
plt.title("Histogram of Food/Feed Industry Output Masses in Tonnes")
plt.xlabel("Tonnes")
plt.ylabel("# of Industries")
```

```python
#Histogram of Item Industry Log Output Mass in 2013
LogItemMass2013 = np.log(IndustryOutput['Y2013']).replace(-np.inf,np.nan).dropna()
#LogItemMass2013
plt.hist(LogItemMass2013)
plt.title("Histogram of Food/Feed Industry Output Masses in Log Tonnes")
plt.xlabel("Log Tonnes")
plt.ylabel("# of Industries")
```

```python
#Checking Normality of Log Distrubtion of Item Industry Mass in 2013 using a QQ Plot
norm1 = np.random.normal(0, 1, len(LogItemMass2013))
norm1.sort()
plt.plot(norm1, LogItemMass2013.sort_values(), 'o')
plt.xlabel("Deliberately Normal Distribution")
plt.ylabel("Log Industry Masses 2013")
plt.title("QQ Plot: Visualizing Normal Distribution of Industry Output Masses")
```

The above QQ plot shows that the Log Food/Feed Industry Mass Outputs in 2013 follow a normal distribution. The QQ plot follows a straight, positively sloping line with the exception of 4 industries: the three smallest industries and the largest industry. Given that these outliers represent the extreme ends of the distribution, it is safe to assume that the Log Masses as a whole is normally distributed.

From the above histogram and QQ plot, we see all the year-to-year global growth percentages follow a normal distribution and that 1992 is an exceptional outlier. Because the distribution is normal, it is possible to perform statistical analysis 

```python
#Countries with Highest Growth Rate in 1992 - Perhaps this will be tell us which countries contributed most to the 1992 global growth rate
CountryOutput[['Y1991','Y1992']].pct_change(axis='columns').sort_values(by='Y1992',ascending=False).replace([np.inf, -np.inf], np.nan).dropna(how='all').head()['Y1992']
```

```python
#To what degree is 1992 an anomoly in yearly global food/feed Growth? (i.e. how many SD's is 1992 food/feed growth percentage away from the mean growth over all years.)
MeanYearGrowth = CountryOutput.sum(axis=0).pct_change(axis='rows').mean()
SDYearGrowth = CountryOutput.sum(axis=0).pct_change(axis='rows').var() ** 0.5
zscore1 = ((MaxYearGrowth - MeanYearGrowth)/SDYearGrowth).values.tolist()[0][0]

print("The Mean Year-to-Year Global Feed/Food Growth Rate between 1961 and 2013 is {}".format(MeanYearGrowth))
print("The SD of these growth rates is {}".format(SDYearGrowth))
print("Therefore the 1991 to 1992 Global Growth rate is {} SD's away from the mean. It is clearly much higher than normal".format(zscore1))
```

# How much livestock does the global feed output produce?

```python
FeedOutput = faodf.loc[faodf['Element'] == 'Feed'].groupby('Item').sum().iloc[:,5:]
FeedOutput.head(5)
```

```python
#Manually check which items count as livestock or livestock derivatives
#faodf['Item'].unique()
#'Bovine Meat', 'Mutton & Goat Meat', 'Poultry Meat', 'Meat, Other', 'Offals, Edible', 'Butter, Ghee',
#'Cream', 'Fats, Animals, Raw', 'Eggs', 'Milk - Excluding Butter', 'Meat', 'Offals', 'Animal fats',
```

```python
FoodOutput = faodf.loc[faodf['Element'] == 'Food']
kwLivestock = ['Meat','Offals','Butter','Cream','Fat','Eggs','Animal fats']
LivestockList = []
```

```python
for kw in kwLivestock:
    for stock in FoodOutput['Item'].unique().tolist():
        if stock not in LivestockList and kw in stock:
            LivestockList.append(stock)
LivestockList
```

```python
LivestockOutput = FoodOutput.groupby('Item').sum().T[LivestockList].T.iloc[:,5:]
LivestockOutput.head(5)
```

```python
LivestockTot = LivestockOutput.sum()
```

```python
#Feed Output shifted forward in time by one year to (arbitrarily) account for lag between production and consumption of feed
#The data set defines feed as the total available amount of feed output available for consumption during the reference period
#Thus the output does not need to be shifted forward by one year to account for consumption.
FeedConsumed = FeedOutput.sum()#.shift()
```

```python
plt.figure(figsize=(15,5))
plt.style.use('seaborn-darkgrid')

plt.subplot(1,2,1)
plt.plot(LivestockTot)
plt.plot(FeedConsumed)
plt.xticks(rotation = 90)

plt.subplot(1,2,2)
plt.plot(LivestockTot/FeedConsumed)
#plt.plot(FeedConsumed/(FeedConsumed + LivestockTot), label='Feed: % of Total Output')
#plt.plot(LivestockTot/(FeedConsumed + LivestockTot), label='LiveStock: % of Total Output')
plt.xticks(rotation = 90)

plt.show()
```
