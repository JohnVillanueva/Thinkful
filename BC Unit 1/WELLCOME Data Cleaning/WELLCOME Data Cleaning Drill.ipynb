```python
wellcome = wellcome.rename(columns={'COST (£) charged to Wellcome (inc VAT when charged)': 'Cost'})
```

```python
upperlist = []
for i in wellcome['Journal title']:
    upperlist.append(str(i).upper())

upperlist    
wellcome['Journal title'] = upperlist

wellcome.head(10)
```

```python
jtitles = sorted(wellcome['Journal title'].dropna().unique().tolist())
jtitles
```

```python
wellcome.groupby('Journal title').count().sort_values(by='Article title', ascending=False)['Article title'].head(5)
```

```python
costlist = []
non_decimal = re.compile(r'[^\d.]+')
for i in wellcome['Cost']:
    costlist.append(float(non_decimal.sub('', i)))
wellcome['COST (£) charged to Wellcome (inc VAT when charged)'] = costlist
```
